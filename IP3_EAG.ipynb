5.	Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip). The data frame will be look like after that

```python

car_sales['Price']=car_sales["Price"].str.replace('$','')
car_sales['Price']=car_sales["Price"].str.replace(',','')
car_sales['Price']
```

```python
car_sales['Price'] = car_sales['Price'].astype(np.int64)
```

```python
car_sales.info()
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function

```python
car_sales['Total Sales']=car_sales['Price'].cumsum(axis=0)
car_sales['Total Sales']
```

7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales). The data frame will be look like that:

```python
salesdate=pd.date_range(start ='2-18-2022', 
         end ='2-27-2022', periods=len(car_sales))
car_sales['Sale Date']=salesdate
```

16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
heat_map=car_sales[['Total Sales','Price', 'Odometer (KM)']].corr(method='pearson')
cols=['Total Sales','Price', 'Odometer (KM)']
axis=sns.heatmap(heat_map, annot=True, xticklabels=cols, yticklabels=cols )
```
